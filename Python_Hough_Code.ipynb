```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np
```

```python
img = cv2.imread('assessment_22.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
blurred_image = cv2.GaussianBlur(img_rgb, (5, 5), 1.4)
```

```python
gray = cv2.cvtColor(blurred_image, cv2.COLOR_RGB2GRAY)
```

```python
edges = cv2.Canny(gray, threshold1=50, threshold2=150)
```

```python
edges_colored = img_rgb.copy()
edges_colored[edges != 0] = [255, 255, 255]
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=50, maxLineGap=10)


if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(edges_colored, (x1, y1), (x2, y2), (0, 255, 0), 2)
```

```python
plt.figure(figsize=(20, 10))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges_colored)
plt.title('Hough Lines on Edges')
plt.axis('off')

plt.show()
```
